```python
df_test = pd.read_feather(r"E:\DS340\MacroHFT\data\ETHUSDT\whole\val.feather")
df_test = df_test.iloc[1:].reset_index(drop=True)
```

```python
import matplotlib.pyplot as plt
action = action.flatten()
df_test['action'] = action

df_test['buy_signal'] = (df_test['action'].shift(1) == 0) & (df_test['action'] == 1)
df_test['sell_signal'] = (df_test['action'].shift(1) == 1) & (df_test['action'] == 0)

# Plotting
fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Plot Price data with buy and sell actions
axs[0].plot(df_test['timestamp'], df_test['close'], label='Price', color='blue')
axs[0].scatter(df_test[df_test['buy_signal']]['timestamp'],
               df_test[df_test['buy_signal']]['close'], color='green', marker='^', label='Buy')
axs[0].scatter(df_test[df_test['sell_signal']]['timestamp'],
               df_test[df_test['sell_signal']]['close'], color='red', marker='v', label='Sell')
axs[0].set_ylabel('Price')
axs[0].legend()
axs[0].grid()

# Assume capital starts at some value and compute capital over time
initial_capital = 100000
capital = initial_capital
shares_held = 0

# Loop through rows to compute capital over time based on buy/sell signals
capital_over_time = []

for index, row in df_test.iterrows():
    if row['buy_signal']:
        # Buy at current price
        if capital > 0:  # Buy only if we have enough capital
            shares_held = capital / row['close']
            capital = 0
    elif row['sell_signal']:
        # Sell at current price
        if shares_held > 0:  # Sell only if we have shares
            capital = shares_held * row['close']
            shares_held = 0
    # Calculate total value (capital + value of shares held)
    total_value = capital + (shares_held * row['close'])
    capital_over_time.append(total_value)

df_test['capital'] = capital_over_time

# Plot Capital over time
axs[1].plot(df_test['timestamp'], df_test['capital'], label='Capital', color='blue')
axs[1].scatter(df_test[df_test['buy_signal']]['timestamp'],
               df_test[df_test['buy_signal']]['capital'], color='green', marker='^', label='Buy')
axs[1].scatter(df_test[df_test['sell_signal']]['timestamp'],
               df_test[df_test['sell_signal']]['capital'], color='red', marker='v', label='Sell')
axs[1].set_ylabel('Capital')
axs[1].set_xlabel('Timestamp')
axs[1].legend()
axs[1].grid()

plt.tight_layout()
plt.show()
```

```python
# Filter data for the first month only
first_month_data = df_test[df_test['timestamp'] < df_test['timestamp'].iloc[0] + pd.Timedelta(days=30)]

# Plotting for the first month only
fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Plot Price data with buy and sell actions for the first month
axs[0].plot(first_month_data['timestamp'], first_month_data['close'], label='Price', color='blue')
axs[0].scatter(first_month_data[first_month_data['buy_signal']]['timestamp'],
               first_month_data[first_month_data['buy_signal']]['close'], color='green', marker='^', label='Buy')
axs[0].scatter(first_month_data[first_month_data['sell_signal']]['timestamp'],
               first_month_data[first_month_data['sell_signal']]['close'], color='red', marker='v', label='Sell')
axs[0].set_ylabel('Price')
axs[0].legend()
axs[0].grid()



plt.tight_layout()
plt.show()
```

```python
second_month_data = df_test[(df_test['timestamp'] >= df_test['timestamp'].iloc[0] + pd.Timedelta(days=30)) &
                            (df_test['timestamp'] < df_test['timestamp'].iloc[0] + pd.Timedelta(days=45))]

# Plotting for the second month only
fig, axs = plt.subplots(2, 1, figsize=(12, 8), sharex=True)

# Plot Price data with buy and sell actions for the second month
axs[0].plot(second_month_data['timestamp'], second_month_data['close'], label='Price', color='blue')
axs[0].scatter(second_month_data[second_month_data['buy_signal']]['timestamp'],
               second_month_data[second_month_data['buy_signal']]['close'], color='green', marker='^', label='Buy')
axs[0].scatter(second_month_data[second_month_data['sell_signal']]['timestamp'],
               second_month_data[second_month_data['sell_signal']]['close'], color='red', marker='v', label='Sell')
axs[0].set_ylabel('Price')
axs[0].legend()
axs[0].grid()

# Plot Capital over time for the second month
axs[1].plot(second_month_data['timestamp'], second_month_data['capital'], label='Capital', color='blue')
axs[1].scatter(second_month_data[second_month_data['buy_signal']]['timestamp'],
               second_month_data[second_month_data['buy_signal']]['capital'], color='green', marker='^', label='Buy')
axs[1].scatter(second_month_data[second_month_data['sell_signal']]['timestamp'],
               second_month_data[second_month_data['sell_signal']]['capital'], color='red', marker='v', label='Sell')
axs[1].set_ylabel('Capital')
axs[1].set_xlabel('Timestamp')
axs[1].legend()
axs[1].grid()

plt.tight_layout()
plt.show()
```
